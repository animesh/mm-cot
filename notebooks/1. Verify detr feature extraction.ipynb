Download the first ScienceQA image.
Apply their DETR feature extraction and verify that the result is equivalent to the features shared as .npy

```python
import urllib
import numpy as np
from src.data.vision_features.detr_extractor import DetrExtractor

extractor = DetrExtractor()

urllib.request.urlretrieve('https://datasets-server.huggingface.co/assets/derek-thomas/ScienceQA/--/derek-thomas--ScienceQA/train/0/image/image.jpg', 'image.jpg')
extractor.extract_vision_features(['image.png'], 'test.npy')

ours = np.load('test.npy')
theirs = np.load('data/vision_features/detr.npy')
```

```python
# Their are almost equal
print(ours[0][0] == theirs[0])

# The difference is at the 4/5th decimal place
for i, _ in enumerate(ours[0][0]):
    for j, _ in enumerate(ours[0][0][i]):
        if ours[0][0][i][j] != theirs[0][i][j]:
            print(ours[0][0][i][j], theirs[0][i][j])
```
